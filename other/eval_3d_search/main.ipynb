```python
import pandas as pd

df = pd.read_csv('random.csv')
first_column = df.iloc[:, 0]

# ids = first_column.head(0) 
ids = first_column
```

```python

import objaverse

def getThumbnail(id):
    annotations = objaverse.load_annotations(uids=[id])
    for annotation in iter(annotations.values()):
            thumbails = annotation["thumbnails"]["images"]    
            first_thumbnail = thumbails[0]
            first_thumbnail_url = first_thumbnail["url"]
            return(first_thumbnail_url)

for id in ids:
    try:
        print(id, getThumbnail(id))
    except Exception:
        pass
```
